# Question 1: How long does it take to stabilize

```python
true_time = list(data[:, TIME])
segments = []
for i, ts in enumerate(true_time):
    if i>=1 and true_time[i-1] > true_time[i]:
        true_time[i] += true_time[i-1]
        
plt.title("time history")
plt.plot(true_time, list(data[:,BEAMCURRENT]))
plt.show()
```

```python

def find_nth_start(n, a):
    # return i
    ctr = 0
    time_list = list(data[:,TIME])
    for i,e in enumerate(time_list):
        if i>=1 and time_list[i-1] > time_list[i]:
            ctr += 1
        if ctr == n:
            return i
    return 0

choose = 14
start = find_nth_start(choose,data)
end = find_nth_start(choose+1,data)
plt.plot(list(data[start:end,TIME]), list(data[start:end,BEAMCURRENT]))
```

```python
for choose in range(10):
    start = find_nth_start(choose,data)
    end = find_nth_start(choose+1,data)
    plt.plot(list(data[start:end,TIME]), list(data[start:end,BEAMCURRENT]), label="segment %d"%choose)
    plt.legend()
```

```python
for choose in range(10,20):
    start = find_nth_start(choose,data)
    end = find_nth_start(choose+1,data)
    plt.plot(list(data[start:end,TIME]), list(data[start:end,BEAMCURRENT]), label="segment %d"%choose)
    plt.legend()
```

```python
for choose in range(20,30):
    start = find_nth_start(choose,data)
    end = find_nth_start(choose+1,data)
    plt.plot(list(data[start:end,TIME]), list(data[start:end,BEAMCURRENT]), label="segment %d"%choose)
    plt.legend()
```

```python
for choose in range(30,40):
    start = find_nth_start(choose,data)
    end = find_nth_start(choose+1,data)
    plt.plot(list(data[start:end,TIME]), list(data[start:end,BEAMCURRENT]), label="segment %d"%choose)
    plt.legend()
```

### Conclusion: In these trials, 30 seconds seems to be enough
